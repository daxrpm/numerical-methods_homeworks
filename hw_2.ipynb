### 3. Encuentre el intervalo m√°s largo en el que se debe encontrar ùëù‚àó para aproximarse a ùëù con error relativo m√°ximo de 10‚àí4 para cada valor de ùëù

```python

def intervalo(p, tol=1e-4):
    delta = tol * abs(p)
    return (p - delta, p + delta)
```

### 4. Use la aritm√©tica de redondeo de tres d√≠gitos para realizar lo siguiente. Calcule los errores absoluto y relativo con el valor exacto determinado para por lo menos cinco d√≠gitos.

```python
def sig_round(x, n=3):
    if x == 0 or math.isnan(x) or math.isinf(x):
        return x
    sign = -1 if x < 0 else 1
    x = abs(x)
    k = int(math.floor(math.log10(x)))
    factor = 10**(n - 1 - k)
    return sign * round(x * factor) / factor
```

```python


def sig_round(x,n=3):
    if x==0: return 0.0
    sign = -1 if x<0 else 1
    x = abs(x)
    k = int(math.floor(math.log10(x)))
    factor = 10**(n-1-k)
    return sign * round(x*factor) / factor

def R(x,n=3): 
    return sig_round(x,n)

x0, y0 = 1.31, 3.24
x1, y1 = 1.93, 5.76
n = 3


t1 = R(R(x0)*R(y1))
t2 = R(R(x1)*R(y0))
num1 = R(t1 - t2)
den = R(R(y1) - R(y0))
x_formula1 = R(num1 / den)


diff = R(R(x1) - R(x0))
num2 = R(diff * R(y0))
x_formula2 = R(R(x0) - R(num2 / den))

x_exact = (x0*y1 - x1*y0)/(y1-y0)

x_formula1, x_formula2, x_exact
```

### Ejercicio 7

Se tienen dos puntos:
$ (x_0, y_0) = (1.31, 3.24), \quad (x_1, y_1) = (1.93, 5.76) $

Con aritm√©tica de redondeo a tres cifras.

---

#### F√≥rmula 1
$$ x = \frac{x_0y_1 - x_1y_0}{y_1 - y_0} $$

$ x_0 y_1 = 1.31 \cdot 5.76 = 7.5456 \Rightarrow 7.55 $  
$ x_1 y_0 = 1.93 \cdot 3.24 = 6.2532 \Rightarrow 6.25 $  
$ 7.55 - 6.25 = 1.30 $  
$ y_1 - y_0 = 5.76 - 3.24 = 2.52 $  
$ x = 1.30 / 2.52 = 0.51587... \Rightarrow \boxed{0.516} $

---

#### F√≥rmula 2
$$ x = x_0 - \frac{(x_1 - x_0)y_0}{y_1 - y_0} $$

$ x_1 - x_0 = 1.93 - 1.31 = 0.62 $  
$ 0.62 \cdot 3.24 = 2.0088 \Rightarrow 2.01 $  
$ 2.01 / 2.52 = 0.7976... \Rightarrow 0.798 $  
$ x = 1.31 - 0.798 = 0.512 \Rightarrow \boxed{0.512} $

---

#### Valor exacto (sin redondeo)
$$ x_{\text{exacto}} = \frac{x_0y_1 - x_1y_0}{y_1-y_0} = 0.5128571428571429 $$
